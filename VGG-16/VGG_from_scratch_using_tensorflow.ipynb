# Import Libraries

```python
import tensorflow
from tensorflow import keras
from keras.layers import Dense,Flatten,MaxPool2D,Input,Conv2D
from keras import Model
```

```python
input = Input(shape = (224,224,3))
```

# VGG 16 Structure

VGG-16.png

# Conv Block 1:
It has two conv layers with 64 filters each,followed by Max Pooling

```python
model = Conv2D(filters = 64 , kernel_size = 3,padding = 'same',activation = 'relu')(input)
model = Conv2D(filters = 54 , kernel_size  = 3,padding = 'same',activation = 'relu')(model)
model = MaxPool2D(pool_size = 2 , strides = 2)(model)
```

# Conv Block 2 :
It has two conv Layers with 128 filters followed by MaxPooling

```python
model = Conv2D(filters  = 128 , kernel_size = 3 , padding = 'same',activation = 'relu')(model)
model = Conv2D(filters = 128 , kernel_size = 3 ,padding = 'same',activation = 'relu')(model)
model = MaxPool2D(pool_size = 2,strides = 2, padding = 'same')(model)
```

# Conv Block 3 :
It has three Conv Layers with 256 filters followed by Max Pooling

```python
model = Conv2D(filters = 256 , kernel_size = 3 , padding = 'same',activation = 'relu')(model)
model = Conv2D(filters = 256,kernel_size = 3, padding = 'same',activation = 'relu')(model)
model = Conv2D(filters = 256,kernel_size = 3, padding = 'same',activation = 'relu')(model)
model = MaxPool2D(pool_size = 2,strides = 2, padding = 'same')(model)
```

# Conv Block 4:
It has three Conv Layer with 512 filters followed by Max Pooling

```python
model = Conv2D(filters = 512 , kernel_size = 3 , padding = 'same',activation = 'relu')(model)
model = Conv2D(filters = 512,kernel_size = 3, padding = 'same',activation = 'relu')(model)
model = Conv2D(filters = 512,kernel_size = 3, padding = 'same',activation = 'relu')(model)
model = MaxPool2D(pool_size = 2,strides = 2, padding = 'same')(model)
```

# Conv Block 5:
It has three Conv Layer with 512 filters followed by Max Pooling

```python
model = Conv2D(filters = 512 , kernel_size = 3 , padding = 'same',activation = 'relu')(model)
model = Conv2D(filters = 512,kernel_size = 3, padding = 'same',activation = 'relu')(model)
model = Conv2D(filters = 512,kernel_size = 3, padding = 'same',activation = 'relu')(model)
model = MaxPool2D(pool_size = 2,strides = 2, padding = 'same')(model)
```

# Dense Layer:
 3 Fully connected Layers , first two layers eith 4096 Hidden units and ReLu activation and the last output layer with 1000 hidden units and softmax activations.

```python
model = Flatten()(model)
model = Dense(units = 4096,activation = 'relu')(model)
model = Dense(units = 4096,activation = 'relu')(model)
output = Dense(units = 1000,activation = 'softmax')(model)
```

# Creating the Model

```python
model = Model(inputs = input , outputs = output)
```

```python
model.summary()
```
